```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [14, 5]
```

#Discretize the continuous signal x(t)=cos(4Ï€t)

```python
time =1
f0=2
kValues =[4,8,10,20]

arrayVector =np.linspace(0,1,1000)
cosineWave = np.cos(4 * np.pi * arrayVector)

plt.figure(figsize=(20, 8))

for index, k in enumerate(kValues, start=1):
    sampleindex = np.linspace(0, time*k*f0, (time*k*f0)+1) #how many smaples = samplesArray
    discreteValues = np.cos((np.pi * sampleindex) / (k/2))


    plt.subplot(2, 2, index)
    plt.stem(sampleindex/(f0*k), discreteValues, label='Samples', use_line_collection=True)
    plt.title(f'k={k}')
    plt.plot(arrayVector, cosineWave, label='Continuous signal')
    plt.legend(loc='lower left') #this is what plot the lengend

plt.show()
```

#Exercise

Considering the Nyquist Theorem, what should be the sampling rate for this signal cos(4*pi*t)? Sample the signal x(t) using the Nyquist sampling rate and plot it. Comment on the result. Remember that: x[n] = x[nTs] = cos[2*pi*fo*n*Ts], that is, t = n.Ts!

```python
#Frequency
f0 = 2
kval = [1,2, 4/3,10]

plt.figure(figsize=(20, 8))
for index, k in enumerate(kval):
  #Sample rate
  ts = 1/(k*f0)

  #samples number
  t = np.arange(0,1+ts,ts)

  #corresponds to each sample point in the time array t
  n = np.arange(0,len(t))

  #here, we're generating the signal
  x = np.cos(2*np.pi*f0*n*ts)

  #plotting the results

  plt.subplot(2, 2, index+1)
  if k == 4/3:
      plt.title(f'k={k} = Aliasing')
  elif k == 1:
      plt.title(f'k={k} = Constant')
  else:
      plt.title(f'k={k}')
  plt.stem(t, x, 'b',use_line_collection=True, label="samples") #ploting the samples
  plt.plot(t, x, '-o', label="creating the wave") #the 'wave'
  plt.legend(loc='lower left') #this is what plot the lengend


plt.show()
```
