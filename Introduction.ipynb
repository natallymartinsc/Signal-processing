#Introduction


```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [14, 5]
```

#Continuous signal

```python
#linspace function -> this function creates a sequence of evenly spaced numbers between two limits
#t is a vector with 10.000 points, starting from 0 and ending at 20
arrayVector = np.linspace(0, 20, 10000) # -> 0 <t <20

#continuous signal calculation
#np.cos(t) -> generates cosine values from each time pont in T = consine wave
continuousSignal = np.cos(arrayVector) * np.exp(-0.1 * arrayVector)

plt.plot(arrayVector, continuousSignal)
```

#Plot of Sine:
The Sine Function has up-down curve which repeats every 2π radians, or 360°.

It starts at 0, heads up to 1 by π/2 radians (90°) and then heads down to −1.

```python
#this function generate an array of 1000 points evenly spaced between -2π and 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
#x is a vector
y_sin = np.sin(x) #sine function of X
y_cos = np.cos(x) #cosine function of x

plt.figure(figsize=(10, 3))
plt.plot(x, y_sin, label='sin(x)', color='green')
plt.title('Sine function')
plt.ylabel('amplitude')
plt.grid(True)
plt.xticks(
    [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    ['-2π', '-3π/2', '-π', '-π/2', '0', 'π/2', 'π', '3π/2', '2π']
)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.ylim(-1.5, 1.5)
plt.legend()
plt.show()

```

#Plot of Cosine
Cosine starts at 1 and heads down until π radians (180°) and then heads up again

```python

plt.figure(figsize=(10, 3))
plt.plot(x, y_cos, label='cos(x)', color='blue')
plt.title('Cosine function')
plt.ylabel('amplitude')
plt.grid(True)
plt.xticks(
    [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    ['-2π', '-3π/2', '-π', '-π/2', '0', 'π/2', 'π', '3π/2', '2π']
)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.ylim(-1.5, 1.5)
plt.legend()

plt.show()

```

#Discrete signal

It's the sampled version of continuous signal

```python
timeSamples = np.linspace(0, 20, 21)
discreteValue = np.cos(timeSamples) * np.exp(-0.1*timeSamples)
#this formula it's to plot the amplitude

plt.stem(timeSamples, discreteValue, label='Discrete Signal', linefmt='b', markerfmt='ro', basefmt=" ")
plt.plot(arrayVector, continuousSignal, '--', label='Continuous Signal')


plt.legend()
plt.show()
```

To plot this graphic, we need calculate for each integer nm with 0<=n<=40 The formula is x[n]= cos(n)e*-0.1*n

```python
#amplitude results
results = {n: np.cos(n) * np.e**(-0.1*n) for n in range(41)}

for n, value in results.items():
    print(f"n={n}: {value}")

```
