```python
from scipy.fft import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (25, 5)
```

```python
# Number of samples points
N = 1000

# Sampling interval
Ts = 1.0 / 1000.0

# Our time vector
x = Ts * np.arange(N)

# A sinusoidal noise signal
noiseFrequency = 0.2 * np.sin(2 * np.pi * x * 100)

# This is our signal composed of two sinusoids
signal = np.sin(10.0 * 2.0 * np.pi * x) + 0.5 * np.sin(30.0 * 2.0 * np.pi * x)

# This is our signal composed of two sinusoids and noise
signalWithNoise = np.sin(50.0 * 2.0 * np.pi * x) + 0.5 * np.sin(80.0 * 2.0 * np.pi * x) + noiseFrequency
```

```python
plt.subplot(1, 3, 1)
plt.title('Noise Frequency')
plt.plot(x, noiseFrequency)

plt.subplot(1, 3, 2)
plt.title('The signal')
plt.plot(x, signal)

plt.subplot(1, 3, 3)
plt.title('Signal with Noise')
plt.plot(x, signalWithNoise)
```

#FFT function

Computes the Fast Fourier Transform of our signal. It transforms the time-domain signal into a freqeuncy-domain representation

```python
yf = fft(signalWithNoise) # It's a complex-valued array representing the amplitude and phase of each freqeuncy

xf = fftfreq(N, Ts)[:N//2] # Generates the freqeuncy bins corresponding to the FFT result
```

np.abs(yf[0:N//2]) - computes the magnitude of the first halp of the FFT

```python
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.title("Magnitude spectrum")

plt.ylabel("Amplitude")
plt.xlabel("Frequency")

plt.grid()
plt.show()
```

```python

plt.plot(xf, np.angle(yf[0:N // 2]))
plt.title('Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')
plt.grid()
```
