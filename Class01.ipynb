#Exercise solving for each k individually

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [14, 5]
```

```python
time= 1
#1 second
f0 = 5
#frequency = 5Hz (Hetz)

#we're creating an array t with 1000 evenly spaced values
#these values start from 0 and go up to time, 1000 it's points in the interval
arrayVector = np.linspace(0, time, 1000)

cosineWave = np.cos(2 * np.pi * arrayVector * f0)

#if k=N=2, So we have 2 samples per complete cycle
k = 2
sampleindex = np.linspace(0, time*k*f0, (time*k*f0)+1)

#print(sampleindex)

#we're calculating the cosine for each n point and we multiplicate by π, generating a discrete signal

#why π? because we're converting to radians, also, it'll calculate the cosine for each sampleIndex
discreteValue = np.cos(np.pi * sampleindex)
#print('samples amplitudes', cosineSample)

plt.subplot(3,2,1) #here we have the grid
#print('Ts', sampleindex/(f0*k))
#Ts is the time of that sample
plt.stem(sampleindex/(f0*k), discreteValue ,use_line_collection=True) #here we're ploting the samples by time and amplitude
plt.plot(arrayVector, cosineWave) #here we're creating the continuous signal
```

#Creating a code to plot for each k value

```python
time = 1
f0=5
kValues = [2,3,4,6] #here, we're setting our k values
arrayVector = np.linspace(0, time, 1000)
cosineWave = np.cos(2 * np.pi * arrayVector * f0) #here we're creating a cosine wave, so it's an array that represents the amplitude for each point in arrayVector

plt.figure(figsize=(20, 8))

for index, k in enumerate(kValues, start=1):
    sampleindex = np.linspace(0, time*k*f0, (time*k*f0)+1)
    discreteValues = np.cos((np.pi * sampleindex) / (k/2))


    plt.subplot(2, 2, index)
    plt.stem(sampleindex/(f0*k), discreteValues, use_line_collection=True)
    plt.plot(arrayVector, cosineWave)

plt.show()
```

#How to create a discrete signal
---
Discretize the continuous signal for x(t)=cos(10πt)






```python
#Here we're creating a function that returns the cos(10πt)
def cosineWaveFunction(time):
    return np.cos(10 * np.pi * time)

arrayVector = np.linspace(0, 1, 1000)  #This array represents a continuous time interval
timeSamples = np.linspace(0, 1, 50)   # contains 50 samples between 0 and 1

# Discrete signal
discreteValue = cosineWaveFunction(timeSamples) #here we're calling the cosineWaveFunction to calculate the discrete signal
#so, foreach t in timeSamples, we're calculating cos(10πt)

# setting the size of the plotting area
plt.figure(figsize=(12, 6))

#continuous signal
plt.plot(arrayVector, cosineWaveFunction(arrayVector), label='continuoues signal $x(t)=\cos(10\pi t)$')

#discrete signal
plt.stem(timeSamples, discreteValue, 'r', markerfmt='ro', basefmt=" ", linefmt='b', label='discrete signal')

plt.title('Discretize the continuous signal $x(t)=\cos(10\pi t)$')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

#Plotting the graphic
plt.show()
```
