#Aliasing

Aliasing is when the sampling rate is too low to capture a high-frequency variation
https://graphics.stanford.edu/courses/

We are using the scipy libary and we are importing three function


*   FFT (Fast Fourier Transform)
*   FFTFREQ: this function generates the frequency values correspongin to the FFT result
*   IFFT (Inverse Fast Fourier Transform):





The Nysquist Frequency is:

f_Nyquist = f_s / 2

The FFT result consists of frequency bins that cover the frequency range from 0 Hz to the Nyquist frequency


```python
from scipy.fft import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 10000
#here is the frequency = 100Hz
#Ts = the sampling interval
Ts = 1.0 / 100.0 # Sampling rate, or number of measurements per second
print(Ts) #this represents 100 Hz
x = Ts*np.arange(N)
#here we have the 4 functions with the 4 frequencies = 25,70,160 e 510

#So we need sum all the sinusoidal components
#the 1.0 value that we multiplicate in every sin function it's the amplitude
signal = 1.0*np.sin(25.0 * 2.0*np.pi*x) + 1.0*np.sin(70.0 * 2.0*np.pi*x) + 1.0*np.sin(160*2.0*np.pi*x) +1.0*np.sin(510*2.0*np.pi*x)

#Fast Fourier Transform and we're passing the Y signal
FFT = fft(signal) #it'll return an array
#the FFT result contains the Frequency Bins which  the frequency bins would cover frequencies from 0 Hz to 50 Hz

#the FFTFREQ we need to pass two arguments: N and Ts
xf = fftfreq(N, Ts)[:N//2]  #it'll generate the frequency value corresponding to the FFT result
print(N//2)


plt.xlabel('Frequency')
plt.ylabel('Magnitude')
#x = hz
plt.plot(xf, 2.0/N * np.abs(FFT[0:N//2]))
plt.grid()
plt.show()
```
