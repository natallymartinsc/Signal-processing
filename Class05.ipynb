#Discrete Convolution

It's a fundamental operation in signal processing, used to combine 2 discrete signals.
Convolution is a mathematical way to combining 2 signals to form a third one, representing how the shape of one is modified by the other.

```python
import numpy as np
from matplotlib import pyplot as plt

# Setting the figure size
plt.rcParams['figure.figsize'] = [12, 5]
```

```python
# Defining our signals
firstSignal = [2, 2, -1]
secondSignal = [4, 3, 2, 1]

# Define the sample indices for the input signals
indexFirstSampleFirstSignal = 0
indexLastSampleFirstSignal = 2

indexFirstSampleSecondSignal = 0
indexLastSampleSecondSignal = 3

# Convolution
convolution = np.convolve(secondSignal, firstSignal)

# Create arrays of sample indices for the input signals
firstSignalIndices = np.arange(indexFirstSampleFirstSignal, indexLastSampleFirstSignal + 1)
secondSignalIndices = np.arange(indexFirstSampleSecondSignal, indexLastSampleSecondSignal + 1)

# Calculate the output indices
# The number of output samples = M + N - 1, where M and N are the lengths of firstSignal and secondSignal respectively
outputIndices = np.arange(min(firstSignalIndices) + min(secondSignalIndices), max(firstSignalIndices) + max(secondSignalIndices) + 1)
```

```python
# Plotting the result
plt.grid("on")
plt.title("Convolution Result")
plt.xlabel("Sample (n)")
plt.ylabel("Amplitude")
plt.stem(outputIndices, convolution, use_line_collection=True)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.show()
```
