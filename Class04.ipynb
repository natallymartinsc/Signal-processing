#Transformation in the independent variable

In signal processing, a transformation in the independent variable refers to the modification or alteration of the time or spatial variable in a signal



*   Time-Scaling: involves compressing or expanding the signal in the time
*   Time-Shifting: involves moving the signal forwar or backward in time

Some real applications:
*   Time-Scaling: used in audio processing to change the speed of playback. In image processing, it can be used to zoom in or out
*   Time-Shifting: used in telecommunications to synchronize signals




```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [12, 5]
```

```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101) #Creating a time vector for continuous signals

#Defining the original function
originalFunction = np.cos(t)*np.exp(-0.1*t)

#Defining the time-reversed function -> x(t) = x(-t)
timeReversedFunction = np.cos(-t)*np.exp(-0.1*-t)

# Creating the original subplot
plt.subplot(1, 2, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, originalFunction, label='Original Function')  # Plotting the original function
plt.title('Original Function')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()  #

# Creating the time-reversed subplot
plt.subplot(1, 2, 2)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, timeReversedFunction, label='Time-Reversed Function', color='orange')  # Plotting the time-reversed function
plt.title('Time-Reversed Function')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()  # Adjusting the layout to avoid overlap
plt.show()
```

#Signal compressing

The signal compressing refers to the process of reducing the amount of data required to represent a signal

```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101) #Creating a time vector for continuous signals

#Defining the original signal
orignalSignal = np.sin(t)*np.exp(-0.1*np.abs(t))

#Defining the compressed signal
a = 2 #this means that each point in time is comrpessed to half its original value
compressedSignal = np.sin(a*t)*np.exp(-0.1*np.abs(a*t))

# Creating the original subplot
plt.subplot(1, 2, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, orignalSignal, label='Original Signal')  # Plotting the original signal
plt.title('Original Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Creating the compressed subplot
plt.subplot(1, 2, 2)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, compressedSignal, label='Compressed Signal', color='orange')  # Plotting the compressed Signal
plt.title('Compressed Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()
plt.show()
```

Ps: When a > 1, the frequency of the signal increases. In our case, the frequency is doubled, so the signal completes its cycles twice as fast.

#Signal Expansion

Signal expansion refers to the process of stretching or expanding the time axis of a signal. This operatoiin slows down a signal, reducing its frequency.


```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101) #Creating a time vector for continuous signals

#Defining the original signal
orignalSignal =np.sin(t)*np.exp(-0.1*np.abs(t))

#Defining the expanded signal
a = 0.5
compressedSignal = np.sin(a*t)*np.exp(-0.1*np.abs(a*t))

# Creating the original subplot
plt.subplot(1, 2, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, orignalSignal, label='Original Signal')  # Plotting the original signal
plt.title('Original Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Creating the expanded subplot
plt.subplot(1, 2, 2)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, compressedSignal, label='Expanded Signal', color='orange')  # Plotting the expanded Signal
plt.title('Expanded Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()
plt.show()
```

PS: When a < 1, the frequency of the signal decreases.

#Signal Advancing

Signal advancingm, also know as time shifting, refers to moving the signal forward in time. This operation shifts the entire signal to the **LEFT** along the time.

If x(t) is our original signal, an advanced signal can be represented as x(t +t0), where t0 is a positive constant. This means that for each point in the original signal is moced t0 units earlier in time

```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101) #Creating a time vector for continuous signals

#Defining the original signal
orignalSignal = np.sin(t)*np.exp(-0.1*np.abs(t))

#Defining the expanded signal
a=1
b = -4
advancedSignal = np.sin(a*t-b)*np.exp(-0.1*np.abs(a*t-b))

# Creating the original subplot
plt.subplot(1, 2, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, orignalSignal, label='Original Signal')  # Plotting the original signal
plt.title('Original Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Creating the advanced subplot
plt.subplot(1, 2, 2)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, advancedSignal, label='Advanced Signal', color='orange')  # Plotting the advanced Signal
plt.title('Advanced Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 1, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, orignalSignal, label='Original Signal')  # Plotting the original signal
plt.plot(t, advancedSignal, label='Advanced Signal', color='orange')  # Plotting the advanced Signal
plt.title('Advanced Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()
plt.show()
```

In practice, advancing a signal is generally not feasible, while delaying a signal is common and pratical operation

#Signal Delay

Delaying a signal, as know as time shifting, is the process of moving the signal to te **RIGHT** along the time axis.

If x(t) is our original signal, an advanced signal can be represented as x(t -t0), where t0 is a positive constant. This means that for each point in the original signal is moced t0 units later in time

```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101) # Creating a time vector for continuous signals

#Defining the original signal
orignalSignal = np.sin(t)*np.exp(-0.1*np.abs(t))

#Defining the delay signal
a = 1
b = 4
delayedSignal = np.sin(a*t-b)*np.exp(-0.1*np.abs(a*t-b))

# Creating the original subplot
plt.subplot(1, 2, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, orignalSignal, label='Original Signal')  # Plotting the original signal
plt.title('Original Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Creating the delayed subplot
plt.subplot(1, 2, 2)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, delayedSignal, label='Delayed Signal', color='orange')  # Plotting the delayed Signal
plt.title('Delayed Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 1, 1)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.grid()
plt.plot(t, orignalSignal, label='Original Signal')  # Plotting the original signal
plt.plot(t, delayedSignal, label='Delayed Signal', color='orange')  # Plotting the advanced Signal
plt.title('Delayed Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Displaying the plots
plt.tight_layout()
plt.show()
```

#Discrete Signal

Discrete signals are defined only at **SPECIFIC INTERVALS OR POINT IN TIME**, unlike continuous signals, which are defined for every instant of time.

Discrete signals are sampled versions of continuou signals and are represented by a sequence of values

```python
plt.rcParams['figure.figsize'] = [12, 5]

# Creating a time vector for continuous signals
t = np.linspace(-10, 10, 101)

# Defining the continuous signal
y_con = np.cos(t) * np.e**(-0.1*t)

n = np.linspace(-10, 10, 41) #Creating a time vector for discrete signals

y_dis = np.cos(n) * np.e**(-0.1*n)

plt.stem(n, y_dis, basefmt=" ", linefmt='b-', markerfmt='bo', label='Discrete Signal') #Plotting the discrete signal

plt.plot(t, y_con, '--', label='Continuous Signal') #Plotting the continuous signal


plt.axvline(0, color='r')
plt.axhline(0, color='r')
# Displaying the plot
plt.grid()
plt.legend()
plt.show()
```
